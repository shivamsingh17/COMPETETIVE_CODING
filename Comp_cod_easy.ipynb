```
# https://www.codechef.com/CJQK2020/problems/CJQK02    given 2 number how many number can be form in which sum is of n number and a is min and b is max
n, a, b=map(int,input().split())

if(n==1):
    if(a == b):
        print(1)
    else:
        print(0)
        
elif(b < a):
    print(0)
else:
    ans = (n-1)*b + a
    t = (n-1)*a + b
    ans = ans - t + 1
    print(ans)
```

```
import scipy.stats as ss
import numpy as np
import matplotlib.pyplot as plt

def plot_beta(x_range, a, b, mu=0, sigma=1, cdf=False, **kwargs):
    '''
    Plots the f distribution function for a given x range, a and b
    If mu and sigma are not provided, standard beta is plotted
    If cdf=True cumulative distribution is plotted
    Passes any keyword arguments to matplotlib plot function
    '''
    x = x_range
    if cdf:
        y = ss.beta.cdf(x, a, b, mu, sigma)
    else:
        y = ss.beta.pdf(x, a, b, mu, sigma)
    plt.plot(x, y, **kwargs)
x = np.linspace(0, 1, 5000)

plot_beta(x, 5, 2, 0, 1, color='red', lw=2, ls='-', alpha=0.5, label='pdf')
plot_beta(x, 5, 2, 0, 1, cdf=True, color='blue', lw=2, ls='-', alpha=0.5, label='cdf')
plt.legend();
```

```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x3 = np.arange(0., 11.,2.)
y3 = x3
x4 = np.arange(0., 11.,1.)
y4 = np.flip(x4,0)

fig=plt.figure(figsize=(8, 6 ))
fig.suptitle('FOUR PLOTS', fontsize=16)

fig.add_subplot(2,2 , 1) 
#1. orange sqaure horizontal line
axes = plt.gca()
axes.set_xlim([0 ,1.0])
axes.set_ylim([0, 1.0])
plt.axhline(y=0.5,xmin=0.,linewidth = 5, xmax=1., color='C1', linestyle=':')

#2. puple dashed purple  line 
fig.add_subplot(2,2 , 2)#2. puple dashed purple  line 
axes = plt.gca()
axes.set_xlim([0 ,1])
axes.set_ylim([0, 1])
plt.axvline(x=0.5 ,ymin=0., ymax=1. ,linewidth= 4, color='C4', linestyle='-.')


#3. blue dashed line
fig.add_subplot(2,2 , 3)
axes = plt.gca()
axes.set_xlim([0 ,10])
axes.set_ylim([0, 10])
plt.plot(x3 ,y3, 'b-.', linewidth= 2)


#4. red dotted line 
fig.add_subplot(2,2 , 4)
axes = plt.gca()
axes.set_xlim([0 ,10])
axes.set_ylim([0, 10])
plt.plot(x4 ,y4, 'ro' )
plt.show()
```
